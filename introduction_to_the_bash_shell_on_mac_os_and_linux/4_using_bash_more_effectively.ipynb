# 5 Pipes

```python
# ls /usr/bin | less  # 把ls命令的输出内容传递到less命令中做进一步交互处理
```

```python
!grep 1978 oscars.tsv
```

```python
!grep 1978 oscars.tsv | sort
```

```python
# 在oscars.tsv中查找包含'1978'的行，将匹配行传递给sort命令排序(默认按字母顺序)，最后将排好序的内容保存到1978_file.txt
!grep 1978 oscars.tsv | sort > 1978_file.txt
```

```python
!cat 1978_file.txt
```

```python
!cut -f 3 oscars.tsv
```

```python
!cut -f 3 oscars.tsv | grep 4
```

```python
# 获取文件中的第三列，筛选包含'4'的行，统计行数
!cut -f 3 oscars.tsv | grep 4 | wc -l
```
